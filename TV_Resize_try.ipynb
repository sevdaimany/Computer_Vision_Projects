</br>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

</br><div align="center"> <b> <font size="3" color='green'>cv2.resize(src, dsize, fx, fy, interpolation)</font></b> </div>
<b>src:</b> (required) source/input image<br>
<b>dsize:</b>(required) desired size for the output image in (width, height) format<br>
<b>fx:</b>(optional) scale factor along the horizontal axis<br>
<b>fy:</b>(optional) scale factor along the vertical axis<br>
<b>interpolation:</b>(optional) could be one of the following values.
 - INTER_NEAREST
 - INTER_LINEAR (used by default)
 - INTER_AREA
 - INTER_CUBIC
 - INTER_LANCZOS4



</br><div align="center"> <b> <font size="4" color='red'>Image resizing with a custom Width and Height</font></b> </div>

```python
image = cv2.imread('pic/Lenna_old.jpg')
```

```python
height, width, channel = image.shape 
print(height, width, channel)
```

```python
heigth_down = 500
width_down = 100
size_down = (width_down, heigth_down)
image_down = cv2.resize(image, size_down)
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1].imshow(cv2.cvtColor(image_down, cv2.COLOR_BGR2RGB))
```

<b> Image.shape return (height, width, channel) but in cv2.resize, the new size determine by (width, height)</b>

</br><div align="center"> <b> <font size="4" color='red'>Resizing With a Scaling Factor</font></b> </div>

```python
scale_x = 2
scale_y = 2
image_scale = cv2.resize(image, None, fx=scale_x, fy=scale_y)
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1].imshow(cv2.cvtColor(image_scale, cv2.COLOR_BGR2RGB))
```

<b> Using None for dsize parameter in mandatory. </b>

</br><div align="center"> <b> <font size="4" color='red'>Resizing With Different Interpolation Methods</font></b> </div>

```python
image = cv2.imread('pic/finger.png')
plt.figure(figsize=(2, 2))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

```python
image_linear_up = cv2.resize(image, None, fx=2, fy=2)
image_near_up = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_NEAREST)
image_cubic_up = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
image_area_up = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_AREA)

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(40, 40))
axes[0].imshow(image_linear_up)
axes[1].imshow(image_near_up)
axes[2].imshow(image_cubic_up)
axes[3].imshow(image_area_up)
```

```python
image_linear_down = cv2.resize(image, None, fx=0.5, fy=0.5)
image_near_down = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_NEAREST)
image_cubic_down = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_CUBIC)
image_area_down = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(5, 5))
axes[0].imshow(image_linear_down)
axes[1].imshow(image_near_down)
axes[2].imshow(image_cubic_down)
axes[3].imshow(image_area_down)
```
