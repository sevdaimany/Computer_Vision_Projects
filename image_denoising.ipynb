```python
import cv2 as cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def image_filter(image,filter):
    f_s = filter.shape[0]
    pad_size = f_s // 2
    image_pading = cv2.copyMakeBorder(image,pad_size,pad_size,pad_size,pad_size,cv2.BORDER_CONSTANT)
    row,col = image_pading.shape
    out_image = np.zeros_like(image)
    for i in range(pad_size,row-pad_size):
        for j in range(pad_size,col-pad_size):
            slice = image_pading[i-pad_size:i+pad_size+1,j-pad_size:j+ pad_size +1]
            out_image[i-pad_size,j-pad_size] = np.sum(slice * filter)
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.imshow(image,cmap='gray')
    plt.subplot(1,2,2)
    plt.imshow(out_image,cmap='gray')
               
            
```

```python
def image_filteriong(image,filter):
    f_s_y = filter.shape[0]
    f_s_x = filter.shape[1]
    pad_size_x = f_s_x //2
    pad_size_y = f_s_y //2
    print(pad_size_x,pad_size_y)
    image_pad = cv2.copyMakeBorder(image,pad_size_y,pad_size_y,pad_size_x,pad_size_x,cv2.BORDER_CONSTANT)
    row,col = image_pad.shape
    out_image = np.zeros_like(image)
    for i in range(pad_size_y,row-pad_size_y):
        for j in range(pad_size_x,col-pad_size_x):
            slice = image_pad[i-pad_size_y:i+pad_size_y+1,j-pad_size_x:j+pad_size_x+1]
            out_image[i-pad_size_y,j-pad_size_x] = np.sum(slice*filter)
    return out_image
```

```python
filter = np.ones((5,5)) / 25
image = cv2.imread('./images/dog.jpeg',0)
image_filter(image,filter)
```
