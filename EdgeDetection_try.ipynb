# Edge detection

</br><div align="center"> <b> <font size="3" color='green'>cv2.Sobel(src, ddepth, dx, dy, ksize)</font></b> </div>
<b>src:</b> (required) source/input image<br>
<b>ddepth :</b>(required) The depth of the image, - 1 indicates that the same depth as the original image is used. The depth of the target image must be greater than or equal to the depth of the original image<br>
<b>dx:</b>(required) The order of derivative in the x-axis direction is generally 0, 1 and 2, where 0 indicates that there is no derivative in this direction<br>
<b>dy:</b>(required) The order of derivative in the y-axis direction is generally 0, 1 and 2, where 0 indicates that there is no derivative in this direction<br>
<b>ksize:</b> Kernel size, default value is 3<br>

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('pic/Arrow.jpg', 0)
img_x = cv2.Sobel(img, -1, 1, 0)
img_y = cv2.Sobel(img, -1, 0, 1)

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,40))
axes[0].imshow(img, cmap='gray')
axes[1].imshow(img_x, cmap='gray')
axes[2].imshow(img_y, cmap='gray')
```

```python
img_x_S = cv2.Sobel(img, cv2.CV_16S, 1, 0)
img_y_S = cv2.Sobel(img, cv2.CV_16S, 0, 1)
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,40))
axes[0].imshow(img, cmap='gray')
axes[1].imshow(img_x_S, cmap='gray')
axes[2].imshow(img_y_S, cmap='gray')
```

```python
img_x_S_np = cv2.convertScaleAbs(img_x_S)
img_y_S_np = cv2.convertScaleAbs(img_y_S)
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,40))
axes[0].imshow(img, cmap='gray')
axes[1].imshow(img_x_S_np, cmap='gray')
axes[2].imshow(img_y_S_np, cmap='gray')
```

## cvtype values


## Gradient magnitude 

```python
img_xy = np.hypot(img_x_S, img_y_S)
plt.imshow(img_xy, cmap='gray')
```

## Laplacian kernel

</br><div align="center"> <b> <font size="3" color='green'>cv2.Laplacian(src, ddepth,ksize)</font></b> </div>
<b>src:</b> (required) source/input image<br>
<b>ddepth :</b>(required) The depth of the image, - 1 indicates that the same depth as the original image is used. The depth of the target image must be greater than or equal to the depth of the original image<br>
<b>ksize:</b> Kernel size, default value is 1<br>

```python
img_lap = cv2.Laplacian(img, cv2.CV_16S,ksize=3)
img_lap_np = cv2.convertScaleAbs(img_lap)
plt.imshow(img_lap_np, cmap='gray')
```

## Sobel and Laplasian comparision

```python
img_lenna = cv2.imread('pic/Lenna.png')
img_lenna_gray = cv2.cvtColor(img_lenna, cv2.COLOR_BGR2GRAY)
img_x_S = cv2.Sobel(img_lenna_gray, cv2.CV_16S, 1, 0)
img_y_S = cv2.Sobel(img_lenna_gray, cv2.CV_16S, 0, 1)
img_xy = np.hypot(img_x_S, img_y_S)
img_lap = cv2.Laplacian(img_lenna_gray, cv2.CV_16S,ksize=3)
img_lap_np = cv2.convertScaleAbs(img_lap)
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,40))
axes[0].imshow(cv2.cvtColor(img_lenna, 4))
axes[1].imshow(img_xy, cmap='gray')
axes[2].imshow(img_lap_np, cmap='gray')
```
