## Read image

</br><div align="center"> <b> <font size="3" color='green'>imread(filename, flags)</font></b> </div>
<b>filename:</b> image path + image name <br>
<b>flags:</b>
 - cv2.IMREAD_UNCHANGED or -1
 - cv2.IMREAD_GRAYSCALE or 0
 - cv2.IMREAD_COLOR or 1

```python
import cv2
img = cv2.imread('pic/Lenna_old.jpg')
height, width, channel = img.shape
print(height, width, channel)
```

Check the result.

```python
if img is None:
    print('image not load')
else:
    print('image load correctly')
```

<b>window_name:</b> A string representing the name of the window in which image to be displayed. </br>

How to add draws and text to images.
- Draw line, shapes (such as cicle, rectangle, ellips).
- Draw filled shapes
- Adding text to the images 

## Draw line
</br><div align="center"> <b> <font size="3" color='green'>line(InputOutput_image, start_point, end_point, color, thickness)</font></b> </div>
<b>InputOutput_image:</b> (required) source image and result image<br>
<b>start_point:</b>(required) starting coordinates of line in 2D (x,y)<br>
<b>end_point:</b>(required) ending coordinates of line in 2D (x,y)<br>
<b>color:</b>(required) color of line to be drawn in BGR <br>
<b>thickness:</b>(optional) thickness of the line in px, default value = 1.


## Draw circle
</br><div align="center"> <b> <font size="3" color='green'>circle(InputOutput_image, center_coordinates, radius, color, thickness)</font></b> </div>
<b>InputOutput_image:</b> (required) source image and result image<br>
<b>center_coordinates:</b> (required) center coordinates of circle in 2D (x,y).</br> 
<b>radius:</b> (required) circle radius. </br> 
<b>color:</b> (required) color of border line of circle to be drawn in BGR</br> 
<b>thickness: </b> thickness of the circle border line in px. Thickness of -1 px will fill the circle shape by the specified color.</br> 

## Draw rectangle

</br><div align="center"> <b> <font size="3" color='green'>rectangle(InputOutput_image, start_point, end_point, color, thickness)</font></b> </div>
<b>InputOutput_image:</b> (required) source image and result image<br>
<b>start_point:</b> (required) starting coordinates of rectangle in 2D (x,y).</br> 
<b>end_point:</b> (required) ending coordinates of rectangle in 2D (x,y).</br> 
<b>color:</b> (required) color of border line of circle to be drawn in BGR</br> 
<b>thickness: </b> thickness of the circle border line in px. Thickness of -1 px will fill the circle shape by the specified color.</br> 

## Draw ellipse

</br><div align="center"> <b> <font size="3" color='green'>ellipse(InputOutput_image, centerCoordinates, axesLength, angle, startAngle, endAngle, color, thickness)</font></b> </div>
<b>InputOutput_image:</b> (required) source image and result image<br>
<b>centerCoordinates:</b> (required) center coordinates of ellipse in 2D (x,y).</br> 
<b>axesLength:</b> (required) tuple of two variables containing major and minor axis of ellipse (major axis length, minor axis length).</br> 
<b>angle:</b> (required)Ellipse rotation angle in degrees. </br>
<b>startAngle:</b> (required) Starting angle of the elliptic arc in degrees.</br> 
<b>endAngle:</b> (required) Ending angle of the elliptic arc in degrees.</br> 
<b>color:</b> (required) color of border line of circle to be drawn in BGR</br> 
<b>thickness: </b> thickness of the border line in px. Thickness of -1 px will fill the circle shape by the specified color.</br> 

```python
img = cv2.imread('pic/flower.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img_ellipse = img.copy()
cv2.ellipse(img_ellipse, (600, 400), (300, 200), 30, 0, 360, (255, 0, 0), 15)
plt.imshow(cv2.cvtColor(img_ellipse, cv2.COLOR_BGR2RGB))
```

## Adding Text

</br><div align="center"> <b> <font size="3" color='green'>putText(InputOutput_image, text, org, font, fontScale, color, thickness)</font></b> </div>
<b>InputOutput_image:</b> (required) source image and result image<br>
<b>text:</b> (required) Text string to be drawn.</br> 
<b>org:</b> (required) Coordinates of the bottom-left corner of the text string in 2D (x,y).</br> 
<b>font:</b> (required)It denotes the font type. Some of font types are FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN, ....</br>
<b>fontScale:</b> (required)  Font scale factor that is multiplied by the font-specific base size.</br> 
<b>color:</b> (required) color of border line of circle to be drawn in BGR</br> 
<b>thickness: </b> thickness of text.</br> 

```python
img_card = cv2.imread('pic/Saman.jpg')
plt.imshow(cv2.cvtColor(img_card, cv2.COLOR_BGR2RGB))
```

```python
img_card_cp = img_card.copy()
numb = '6037 9974 3833 0244'
cv2.putText(img_card_cp, numb, (100, 200), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 2)
plt.imshow(cv2.cvtColor(img_card_cp, cv2.COLOR_BGR2RGB))
```
