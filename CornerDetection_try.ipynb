# Corner Detection

</br><div align="center"> <b> <font size="3" color='green'>cv2.cornerHarris(src, blockSize, ksize, k)</font></b> </div>
<b>src: </b>Input single-channel 8-bit or floating-point image. </br>
<b>blockSize:</b>Neighborhood size. </br>
<b> ksize: </b>Aperture parameter for the Sobel operator.</br>
<b>k: </b>Harris detector free parameter. See the formula above (0.04-0.06).</br>

<b>dst: </b>	Image to store the Harris detector responses. It has the type CV_32FC1 and the same size as src .

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('PartLogo.png')
plt.figure(figsize=(10,20))
plt.imshow(cv2.cvtColor(img, 4))
```

```python
img_corner = img.copy()
img_gray = cv2.cvtColor(img, 6)
img_gray = np.float32(img_gray)
corners = cv2.cornerHarris(img_gray, 2, 3, 0.04)
corners = cv2.dilate(corners,None)
img_corner[corners>0.5*corners.max()]=[0,0,255]
plt.figure(figsize=(20,40))
plt.imshow(cv2.cvtColor(img_corner, 4))
```

</br><div align="center"> <b> <font size="3" color='green'>cv2.cornerMinEigenVal(src, blockSize, ksize)</font></b> </div>
<b>src: </b>Input single-channel 8-bit or floating-point image. </br>
<b>blockSize:</b>Neighborhood size. </br>
<b> ksize: </b>Aperture parameter for the Sobel operator.</br>

<b>dst: </b>	Image to store the Harris detector responses. It has the type CV_32FC1 and the same size as src .

```python
img_corner = img.copy()
img_gray = cv2.cvtColor(img, 6)
img_gray = np.float32(img_gray)
corners = cv2.cornerMinEigenVal(img_gray, 2, 3)
corners = cv2.dilate(corners,None)
img_corner[corners>0.05*corners.max()]=[0,0,255]
plt.figure(figsize=(20,40))
plt.imshow(cv2.cvtColor(img_corner, 4))
```

</br><div align="center"> <b> <font size="3" color='green'>cv2.goodFeaturesToTrack(gray_img, maxc, Q, minD)</font></b> </div>
<b>gray_img: </b>Input single-channel 8-bit or floating-point image. </br>
<b>maxc:</b>Maximum number of corners we want(give negative value to get all the corners) </br>
<b>Q: </b>Quality level parameter(preferred value=0.01)</br>
<b>minD: </b>Maximum distance(preferred value=10)</br>


```python
img_corner = img.copy()
img_gray = cv2.cvtColor(img, 6)
img_gray = np.float32(img_gray)
corners = cv2.goodFeaturesToTrack(img_gray, 10, 0.01, 10)
corners = np.int0(corners)
print(corners)
for corner in corners:
    cv2.circle(img_corner, corner[0], (0,0,255), 3, -1)
    
plt.imshow(img_corner)
```
