<b>src:</b> It is the image whose color space is to be changed.</br>
<b>code:</b> It is the color space conversion code. </br>

 - cv2.COLOR_BGR2RGB = 4,
 
 - cv2.COLOR_BGR2GRAY = 6,
 
 - ...



```python
import cv2
import matplotlib.pyplot as plt

img_indoor = cv2.imread('pic/cube_indoor.png')
img_outdoor = cv2.imread('pic/cube_outdoor.png')

fig, sub = plt.subplots(nrows=1, ncols=2)
sub[0].imshow(cv2.cvtColor(img_indoor, 4))
sub[1].imshow(cv2.cvtColor(img_outdoor, 4))
```

```python
def show_results(img, img_c):
    img_c_ch1, img_c_ch2, img_c_ch3 = cv2.split(img_c)
    fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 20))
    axes[0].imshow(cv2.cvtColor(img, 4))
    axes[1].imshow(cv2.cvtColor(img_c_ch1, 4))
    axes[2].imshow(cv2.cvtColor(img_c_ch2, 4))
    axes[3].imshow(cv2.cvtColor(img_c_ch3, 4))
```

## RGB Model

```python
img_indoor_RGB = cv2.cvtColor(img_indoor, 4)
show_results(img_indoor, img_indoor_RGB)
```

```python
img_outdoor_RGB = cv2.cvtColor(img_outdoor, 4)
show_results(img_outdoor, img_outdoor_RGB)
```

## HSV Model

```python
img_indoor_HSV = cv2.cvtColor(img_indoor, cv2.COLOR_BGR2HSV)
show_results(img_indoor, img_indoor_HSV)
```

```python
img_outdoor_HSV = cv2.cvtColor(img_outdoor, cv2.COLOR_BGR2HSV)
show_results(img_outdoor, img_outdoor_HSV)
```

## LAB Model

```python
img_indoor_LAB = cv2.cvtColor(img_indoor, cv2.COLOR_BGR2LAB)
show_results(img_indoor, img_indoor_LAB)
```

```python
img_outdoor_LAB = cv2.cvtColor(img_outdoor, cv2.COLOR_BGR2LAB)
show_results(img_outdoor, img_outdoor_LAB)
```

<b>src:</b> input image</br>
<b>lowerb:</b> lower boundary array</br>
<b>upperb:</b> upper boundary array</br>

## Green space detection using color thresholding

```python
img_city = cv2.imread('pic/city.JPG')
img_city_lab = cv2.cvtColor(img_city, cv2.COLOR_BGR2LAB)
l , a, b = cv2.split(img_city_lab)
print('lower bound', l.min(), a.min(), b.min())
print('upper bound', l.max(), a.max(), b.max())
show_results(img_city, img_city_lab)
```

```python
lower_bound = (0, 0, 150)
upper_bound = (255,150,255)
img_city_greenspace = cv2.inRange(img_city_lab, lower_bound, upper_bound)
plt.figure(figsize=(5,5))
plt.imshow(img_city_greenspace, cmap='gray')
print(sum(sum(img_city_greenspace==255)))
```
